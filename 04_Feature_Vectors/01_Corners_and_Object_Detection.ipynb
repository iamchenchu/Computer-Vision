## Introduction to ORB (Oriented Fast and Rotated Brief)

* ORB, starts by finding special regions in an image called key points.
* __Keypoint :__ a small region in an image that is particularly distinctive. For example corners where pixel values sharply chnage from light to dark
    
  
* ORB calculate the corresponding the feature vector once it locates the keypoints in the image.
* All the feature vectors will be in in the binary represetion for ORB, so these are called Binary Feature Vectors
* ORB also imrpovious to noise illumination, and image transformations such as rotations.


### FAST Algorithm

* The first step in ORB feature detection is to find the key points in an image, which is done by the FAST algorithm 
* FAST : Features from Accelerated Segments Test, it quickly selects the key points by comparing the brightness levels in the given pixel area.
  
* FAST compares the brightness of a p area to a set of 16 surrounding pixels that are in a small circule arround p. Each pixel will be sorted to 3 classes
      + Brighter than p
      + Darker than p
      + Similar to p
* Once it compares the p with the classes it will give us the key points but it won't give us the direction of the change of the pixels

### BRIEF
* To create the feature vectors from the keypoints the ORB usees the BRIEF algorithm.
* BRIEF : Binary Robust Independent Elementary Features and it's purpose is to create the binary vectors from the set of key points
* The great advantage of the storing the vectors in the binary form is that they can be stored very efficiently and can be calculated very quickly.
* These properties not only make BRIEF very fast, which is crutial for real time applications. But also allow BRIEF to run on device with very limited computational resources.
