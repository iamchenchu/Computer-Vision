__Regularization__ in deep learning refers to techniques used to prevent __overfitting by adding constraints or penalties to the model__ during training. Overfitting occurs when a model performs well on training data but poorly on unseen data due to excessive complexity or learning noise rather than the underlying pattern. Regularization helps in improving the generalization capability of the model.

__L1 and L2 Regularization:__

__L1 Regularization (Lasso):__ Adds the absolute value of the magnitude of coefficients as a penalty term to the loss function. It encourages sparsity in the model weights, effectively leading some weights to become zero, which can be useful for feature selection.

__L2 Regularization (Ridge):__ Adds the squared magnitude of coefficients as a penalty term to the loss function. It discourages large weights and helps in smoothing the model, which often improves its generalization ability.

# Local Minima

local minima refer to points in the parameter space where the objective function (typically the loss function in neural networks) has a lower value than at nearby points but is not necessarily the lowest value possible (which would be the global minimum). In other words, a local minimum is a point where the function has a minimum value within a surrounding region, but there may exist other regions where the function takes an even lower value.

## Batch vs Stochastic Gradient Descent

To reach we need to certain point we will take the number of steps in the descent process. Here in this case the number of steps means the number of epochs so be mindful about it.

Gradient Descent is a fundamental optimization algorithm used to minimize the loss function in machine learning and deep learning models by iteratively updating the model's parameters. There are several variants of gradient descent, with Batch Gradient Descent and Stochastic Gradient Descent (SGD) being two of the most commonly used approaches.

Batch Gradient Descent computes the gradient of the loss function with respect to the model parameters by using the entire training dataset. In each iteration (or epoch), the model is updated based on the gradient calculated from all data points.

Stochastic Gradient Descent updates the model parameters for each individual training example. Instead of computing the gradient over the entire dataset, it computes the gradient and updates the parameters after evaluating each data point.
