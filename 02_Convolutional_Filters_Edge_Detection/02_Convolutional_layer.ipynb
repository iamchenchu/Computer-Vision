## Canny Edge detection

So far we have seen the images detecting edges using lowpass , highpass filters and also binary threshold as below 


but we need to know what level of intensity chnage constitutes an edge ?
and also we should think of how we can consistantly represent thin and thick edges ?

One of the best answer for the above problems is `Canny Edge Detection`
 



## Canny Edge Detection

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

#Read in the image
image = cv2.imread("images/brain_MR.jpg")

# Change color to RGB (from BGR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) 

plt.imshow(image)
```

```python
# Convert the image to grayscale for processing
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

### Implement Canny edge detection

```python
# Try Canny using "wide" and "tight" thresholds

wide = cv2.Canny(gray, 30, 100)
tight = cv2.Canny(gray, 200, 240)

#Display the images
f, ax = plt.subplots(1, 2, figsize=(10, 10))

ax[0].set_title("wide")
ax[0].imshow(wide, cmap='gray')

ax[1].set_title("tight")
ax[1].imshow(tight, cmap='gray')
```

```python
# Read in the image
image = cv2.imread('images/sunflower.jpg')

# Change color to RGB (from BGR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python
# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

## TODO: Define lower and upper thresholds for hysteresis
# right now the threshold is so small and low that it will pick up a lot of noise
lower = 120
upper = 240

edges = cv2.Canny(gray, lower, upper)
plt.imshow(edges, cmap='gray')
```

```python
# Try Canny using 'wide' and 'tight' thresholds

wide = cv2.Canny(gray, 30, 100)
tight = cv2.Canny(gray, 180, 240)

# Display the images
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 7))

ax1.set_title('wide')
ax1.imshow(wide, cmap='gray')

ax2.set_title('tight')
ax2.imshow(tight, cmap='gray')
                             
```
